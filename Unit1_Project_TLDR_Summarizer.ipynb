```python
from transformers import pipeline
```

```python
# # Load Summarization Model
# summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
text = """
Generative Artificial Intelligence is a rapidly evolving field that focuses on creating systems capable of generating original content, including written text, images, music, videos, and software code. By training on massive datasets, these models learn underlying patterns and relationships, allowing them to produce outputs that closely mimic human creativity. Generative AI is increasingly used in applications such as virtual assistants, automated content writing, language translation, summarization, and software development. Despite its advantages, the technology raises concerns related to inaccurate outputs, algorithmic bias, data privacy, and the misuse of AI-generated media. As a result, responsible development, transparency, and ethical guidelines are essential to ensure safe and trustworthy use of generative AI technologies.
"""
```

```python
# # generate summary

summary = summarizer(text, max_length=30, min_length=10, do_sample=False)
print(summary[0]['summary_text'])

```
